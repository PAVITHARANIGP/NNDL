### AND GATE PERCEPTRON IMPLEMENTAION



```python
# IMPORTING THE REQUIRED MODULE IN THE WORK SPACE
import numpy as np
import matplotlib.pyplot as plt

# AND GATE TRUTH TABLE
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# ACTIVATION FUNCTION DEFINITION
def step_function(x):
    return 1 if x >= 0 else 0

# PERCEPTRON CLASS
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return step_function(summation)

    def train(self, X, y):
        for _ in range(self.epochs):
            for inputs, label in zip(X, y):
                prediction = self.predict(inputs)
                error = label - prediction
                self.weights += self.learning_rate * error * inputs
                self.bias += self.learning_rate * error

# CREATING AND TRAINING PERCEPTRON
perceptron = Perceptron(input_size=2)
perceptron.train(X, y)

# TESTING
print("AND GATE RESULTS:")
for inputs in X:
    prediction = perceptron.predict(inputs)
    print(f"INPUT: {inputs}, PREDICTED OUTPUT: {prediction}")

print(f"FINAL WEIGHTS: {perceptron.weights}")
print(f"FINAL BIAS: {perceptron.bias}")

# DECISION TREE BOUNDARY PLOT
plt.scatter(X[:, 0], X[:, 1], c=y)
x1 = np.linspace(-0.5, 1.5, 100)
x2 = -(perceptron.weights[0] * x1 + perceptron.bias) / perceptron.weights[1]
plt.plot(x1, x2, 'r-')
plt.xlabel('INPUT 1 ')
plt.ylabel('INPUT 2 ')
plt.title('AND GATE - DECISION BOUNDARY')
plt.show()
```

## QUESTIONS
### Q1. How do the weights and bias values change during training for the AND gate?

*The weights and bias start with random values and are adjusted during training. For the AND gate, you'll typically see positive weights for both inputs and a negative bias. This allows the perceptron to output 1 only when both inputs are 1*

### Q2. Can the perceptron successfully learn the AND logic with a linear decision boundary?

*Yes, the perceptron can successfully learn the AND logic with a linear decision boundary. The AND function is linearly separable, which means a single line (in 2D space) can separate the positive case (1) from the negative cases (0)*

### XOR GATE IMPLEMENTATION

### Q1. Why does the Single Layer Perceptron struggle to classify the XOR gate?

*The XOR function is not linearly separable. This means that no single straight line can be drawn to separate the positive cases (1) from the negative cases (0) in the input space. A Single Layer Perceptron can only create linear decision boundaries, which is insufficient for the XOR problem.*

### Q2. What modifications can be made to the neural network model to handle the XOR gate correctly?

*To correctly classify the XOR gate, we need to use a multi-layer perceptron (MLP) with at least one hidden layer. This allows the network to create non-linear decision boundaries. A common architecture for solving XOR is a network with two neurons in the hidden layer and one output neuron. This setup can create the necessary non-linear decision boundary to separate the XOR inputs correctly.*
